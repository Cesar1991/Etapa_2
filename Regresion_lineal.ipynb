```python
# Importar las bibliotecas necesarias

import matplotlib.pyplot as plt
from sklearn import linear_model
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
datos =pd.read_csv("./data.csv",sep=",")
```

```python
datos
```

```python
#definicion input y output

X_datos = np.array(datos["metro"]).reshape((-1,1))
Y_datos = np.array(datos["precio"])

# Crear un modelo de regresi√≥n lineal
model = linear_model.LinearRegression(fit_intercept = False)
model.fit(X_datos, Y_datos)


#imprimiendo parametros
b = model.intercept_
m = model.coef_

print(f"intercepto (b): {b}")
print(f"pendiente (m): {m}")

#predicciones
predicciones = model.predict(X_datos)
predicciones


```

```python
# Graficar los datos y la l√≠nea de regresi√≥n
plt.scatter(X_datos, Y_datos, label='Datos originales', color='blue')
plt.scatter(X_datos, predicciones, label='Predicciones', color='red')
plt.plot(X_datos, predicciones, color='black', label='L√≠nea de Regresi√≥n')
plt.xlabel('Metro')
plt.ylabel('Precio')
plt.legend()
# Agregar etiquetas con coordenadas en cada punto
for i, (x, y) in enumerate(zip(X_datos, Y_datos)):
    plt.annotate(f'({x}, {y})', (x, y), textcoords='offset points', xytext=(0, 11), ha='right')

for i, (x, y) in enumerate(zip(X_datos, predicciones)):
    plt.annotate(f'({x}, {y})', (x, y), textcoords='offset points', xytext=(0, 1), ha='left')

plt.show()
```
